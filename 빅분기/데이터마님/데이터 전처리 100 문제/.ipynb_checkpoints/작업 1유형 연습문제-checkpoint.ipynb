# 4번 일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?

```python
df
```

```python
df.info()
```

```python
df.loc[:,'trending_date2'] = pd.to_datetime(df['trending_date2'])
```

```python
df.info()
```

```python
df.loc[df['trending_date2'].dt.day_name() == 'Sunday']['categoryId'].value_counts()
```

```python
df.loc[df['trending_date2'].dt.day_name() == 'Sunday']['categoryId'].value_counts().index
```

```python
answer = df.loc[df['trending_date2'].dt.day_name() == 'Sunday']['categoryId'].value_counts().index[0]
print(answer)
```

# 5번 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라

```python
df
```

```python
group = df.groupby([df['trending_date2'].dt.day_name(),'categoryId'],as_index= False).size()
group
```

```python
group.pivot(index='categoryId',columns='trending_date2')
```

# 6번 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다. viewcount대비 댓글수가 가장 높은영상을 확인하라 (view_count값이 0인 경우는 제외한다)

```python
df
```

```python
data = df.loc[df['view_count']!= 0]
data
```

```python
data['ratio'] = (data['comment_count']/data['view_count'])
```

```python
data
```

```python
data.sort_values(by = 'ratio',ascending = False)
```

```python
data.sort_values(by = 'ratio',ascending = False).iloc[0,]
```

```python
data.sort_values(by = 'ratio',ascending = False).iloc[0,]['title']
```

# 7번 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)

```python
df
```

```python
(df['comment_count'] / df['view_count']).dropna().sort_values()
```

```python
ratio = (df['comment_count'] / df['view_count']).dropna().sort_values()
ratio[ratio!=0]
```

```python
ratio[ratio!=0].index[0]
```

```python
df.iloc[ratio[ratio!=0].index[0]]
```

```python
result= df.iloc[ratio[ratio!=0].index[0]]['title']
print(result)
```

# 8번 like 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)

```python
df
```

```python
data = df.loc[(df['likes'] != 0) & (df['dislikes'] != 0)]
data
```

```python
num = (data['dislikes']/data['likes']).sort_values().index[0]
num
```

```python
df.iloc[num]['title']
```
