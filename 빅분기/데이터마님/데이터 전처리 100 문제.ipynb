# 24번 item_price 컬럼의 달러표시 문자를 제거하고 float 타입으로 저장하여 new_price 컬럼에 저장하라

```python
a = '$123.23'
a.replace('$','')
```

```python
df['item_price']
```

```python
df['item_price'].str.replace('$','')
# .str.replace : 데이터 시리즈의 값들중에서 일부요소만 바꾸고 싶으면 앞에 .str을 붙여야한다.
# .str없이 그냥 .replace를 사용하면 바꾸고 싶은 기존값이 입력값이 정확히 동일해야 한다.
```

```python
df['item_price'].str.replace('$','',regex = False)
# $가 정규 표현식에 사용 되는 문자이기 때문에 regex = False를 사용
#regex = False 정규표현식 인식여부 설정, 기입을 하지않으면 버전이 다르거나 상황에 따라 에러가 발생할수도있다.
```

```python
df['item_price'].str.replace('$','',regex = False).astype(float) #.astype() : 원하는 타입으로 변경
```

```python
df['new_price'] = df['item_price'].str.replace('$','',regex = False).astype(float)
```

```python
df.loc[(df['item_name']=='Steak Salad') | (df['item_name']=='Bowl')].drop_duplicates('item_name')
#DataFrame.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
#drop_duplicates메서드는 내용이 중복되는 행을 제거하는 메서드
#subset : 중복값을 검사할 열 입니다. 기본적으로 모든 열을 검사합니다.
#keep : {first / last} 중복제거를할때 남길 행입니다. first면 첫값을 남기고 last면 마지막 값을 남깁니다.
#inplace : 원본을 변경할지의 여부입니다.
#ignore_index : 원래 index를 무시할지 여부입니다. True일 경우 0,1,2, ... , n으로 부여됩니다.
```
