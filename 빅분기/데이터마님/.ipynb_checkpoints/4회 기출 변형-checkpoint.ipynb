예측 변수 Segmentation

test.csv에 대해 ID별로 Segmentation의 클래스를 예측해서 저장후 제출
<br>
제출 데이터 컬럼은 ID와 Segmentation 두개만 존재해야함.
<br>
평가지표는 macro f1 score

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e4_p2_test.csv')

display(train.head(),test.head())
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
na_dic= {
    'Ever_Married':train['Ever_Married'].value_counts().index[0],
    'Graduated' : train['Graduated'].value_counts().index[0],
    'Profession' : train['Profession'].value_counts().index[0],
    'Work_Experience' : train['Work_Experience'].mean(),
    'Family_Size' : train['Family_Size'].value_counts().index[0],
    'Var_1' : train['Var_1'].value_counts().index[0],
}    
```

```python
for i in na_dic.keys():
    train[i] = train[i].fillna(na_dic[i])
    test[i] = test[i].fillna(na_dic[i])
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train['Segmentation'].unique()
```

```python
y = train['Segmentation']
y
```

```python
x = train.drop(columns=['ID','Segmentation'])
x
```

```python
test_x = test.drop(columns=['ID'])
test_x
```

```python
x_dummies = pd.get_dummies(x)
x_dummies
```

```python
x_test_dummldes = pd.get_dummies(test_x)[x_dummies.columns]
x_test_dummldes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_valid, Y_train, Y_valid = train_test_split(x_dummies, y, random_state=42, stratify=y)
```

train_test_split(stratify=None)
<br>
stratify = 타겟data로 지정을 해주면 클래스가 비율에 따라 분포시켜준다.
<br>
https://yeko90.tistory.com/entry/what-is-stratify-in-traintestsplit

```python
rf = RandomForestClassifier(random_state=42)
```

```python
rf.fit(X_train, Y_train)
```

```python
predict_valid_label = rf.predict(X_valid)
predict_valid_label
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(Y_valid,predict_valid_label, average='macro')
```

```python
predict_test_label = rf.predict(x_test_dummldes)
predict_test_label
```

```python
pd.DataFrame({'ID':test['ID'],'Segmentation':predict_test_label})
```

## 검정 통계량을 소숫점 이하 3자리에서 구하시오

```python
import numpy as np
n = 100 # 표본 수
p_hat = 0.6 # 표본비율
p = 0.5 # 검정 비율

# 검정 통계량 계산 : (표본 비율 - 검정비율) / 루트(검점비율(1-검정비율) / 표본 수)
z = round((p_hat - p) / np.sqrt(p * (1 - p) / n),3)
print(z)
```

## pvalue를 소숫점 이하 3자리까지 구하고 귀무가설과 대립가설중 유의한 것을 출력하라

```python
from scipy.stats import norm # 정규분포
p_value = round(1 - norm.cdf(z),3) # 1-a
p_value
```

```python
if p_value >= 0.5:
    print('귀무')
elif p_value < 0.5:
    print('대립')
```
