```python
end = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') =='2021-11-01 15'][['channelname','subcnt']].reset_index(drop =True)
end.columns = ['channelname','end_subcnt']
end
```

```python
sub = pd.merge(start,end)
sub
```

```python
sub['del'] = sub['end_subcnt']-sub['start_subcnt']
sub
```

```python
answer = sub.loc[:,['channelname','del']]
answer
```

# 19번 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오

```python
video
```

```python
up = video[(video['ct'].dt.strftime('%Y-%m-%d %H:%M:%S') >= '2021-11-01 00:00:00') & \
      (video['ct'].dt.strftime('%Y-%m-%d %H:%M:%S') <= '2021-11-01 15:00:00')]
up
```

```python
# groupby를 집계할 때, agg()를 이용하면 다수의 함수를 한번에 적용
def sub (x):
    result = max(x)-min(x)
    return result
```

```python
up[['videoname','viewcnt']].groupby('videoname').agg(sub)
```

# 56번 Temperature의 값이 10미만의 경우 A, 10이상 20미만의 경우 B,20이상 30미만의 경우 C, 그 외의 경우 D라고 할때 각 단계의 데이터 숫자를 구하여라

```python
def Tp (x):
    if x < 10:
        return 'A'
    elif x >= 10 and x <20:
        return 'B'
    elif x >= 20 and x <30:
        return 'C'
    elif x >= 30:
        return 'D'
```

# 75번 남성 중 A등급과 D등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라

```python
df.head()
```

```python
df['측정회원성별'].unique()
```

```python
df1 = df[df['측정회원성별'] == 'M']
df1
```

```python
df2 = df1[df1['등급'].isin(['A','D'])]
df2
```

```python
df2.groupby('등급').mean()['체지방율 : %']
```

```python
abs(df2.groupby('등급').mean()['체지방율 : %'][0] - df2.groupby('등급').mean()['체지방율 : %'][1])
```

## 7번 bmi는 자신의 몸무게(kg)를 키의 제곱(m)으로 나눈값이다. 데이터의 bmi 를 구한 새로운 컬럼을 만들고 남성의 bmi 평균을 구하여라

```python
df.head()
```

```python
df['bmi'] = df['체중 : kg'] / ((df['신장 : cm'] / 100) ** 2)
df.head()
```

```python
df[df['측정회원성별'] == 'M']['bmi'].mean()
```

# 78번 bmi보다 체지방율이 높은 사람들의 체중평균을 구하여라

```python
df.head()
```

```python
df[df['bmi'] < df['체지방율 : %']]['체중 : kg'].mean()
```
