```python
answer = sub.loc[:,['channelname','del']]
answer
```

# 19ë²ˆ 2021-11-01 00:00:00 ~ 15:00:00ê¹Œì§€ ê° ì—í”¼ì†Œë“œë³„ viewcntì˜ ì¦ê°€ëŸ‰ì„ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ë§Œë“œì‹œì˜¤

```python
video
```

```python
up = video[(video['ct'].dt.strftime('%Y-%m-%d %H:%M:%S') >= '2021-11-01 00:00:00') & \
      (video['ct'].dt.strftime('%Y-%m-%d %H:%M:%S') <= '2021-11-01 15:00:00')]
up
```

```python
# groupbyë¥¼ ì§‘ê³„í•  ë•Œ, agg()ë¥¼ ì´ìš©í•˜ë©´ ë‹¤ìˆ˜ì˜ í•¨ìˆ˜ë¥¼ í•œë²ˆì— ì ìš©
def sub (x):
    result = max(x)-min(x)
    return result
```

```python
up[['videoname','viewcnt']].groupby('videoname').agg(sub)
```
