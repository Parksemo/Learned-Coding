DataUrl = https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3.csv

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3.csv')
df
```

## **var_6 컬럼의 1,3사분위수 각각 IQR의 1.5배 벗어난 이상치의 숫자를 구하라 **

```python
# IQR : 3분위수 - 1분위수
# 이상치 : 최소값(1분위수 - 1.5*IQR)보다 작거나 최대값(3분위수 + 1.5*IQR)보다 큰 수치들
```

```python
IQR = df['var_6'].quantile(0.75) - df['var_6'].quantile(0.25)
IQR
```

```python
min_value = df['var_6'].quantile(0.25) - (1.5 * IQR)
min_value
```

```python
max_value = df['var_6'].quantile(0.75) + (1.5 * IQR)
max_value
```

```python
len(df[(df['var_6'] > max_value) | (df['var_6'] < min_value)])
```

<br>
종속변수 :이용금액 , 평가지표 : rmse

```python
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_train2.csv')
test = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p2_test2.csv')

display(train.head(),test.head())
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train.nunique()
```

```python
test.nunique()
```

```python
x = train.drop(columns=['ID','이용금액'])
```

```python
test_x = test.drop(columns=['ID'])
```

```python
x_dummies = pd.get_dummies(x)
```

```python
x_test_dummies = pd.get_dummies(test_x)
```

```python
x_test_dummies = x_test_dummies[x_dummies.columns]
```

```python
y = train['이용금액']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_valid, Y_train, Y_valid = train_test_split(x_dummies,y,random_state = 42) #회귀문제라 stratify사용 안하는 게 맞는듯?
```

```python
rf = RandomForestRegressor(random_state = 42)
```

```python
rf.fit(X_train, Y_train)
```

```python
predict_valid = rf.predict(X_valid)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
```

```python
np.sqrt(mean_squared_error(Y_valid,predict_valid))
```

```python
predict_test = rf.predict(x_test_dummies)
```

```python
pd.DataFrame({'ID':test['ID'],'이용금액':predict_test})
```

종속변수 : Target

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3_1.csv')
df
```

## 선형관계 가장 큰 변수 찾아 상관계수를 구하여라

```python
df.corr()['Target'].sort_values(ascending = False).values[1]
```

## Target 변수를 종속변수로 하여 다중선형회귀모델링을 진행했을 때 v2 컬럼의 회귀 계수는?

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(df.drop(columns = ['Target']))
X
```

```python
s = sm.OLS(df['Target'],X)
```

```python
ss = s.fit()
```

```python
ss.summary()
```

```python
ss.params['v2']
```

## 회귀 계수들이 가지는 p값들 중 최대 값은?

```python
ss.pvalues.sort_values(ascending = False)
```

```python
ss.pvalues.max()
```

심장병 발병 예측 / 종속변수 target 1: 발병
<br>
데이터 출처 : uci머신러닝
<br>
데이터 url : https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3_t.csv
<br>
train 데이터는 앞의 210개 행을, test데이터는 나머지 부분을 사용한다

```python
import pandas as pd 
df= pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/krdatacertificate/e7_p3_t.csv')
df
```

## train 데이터로 target을 종속변수로 로지스틱 회귀를 진행할 때 age 컬럼의 오즈비를 구하여라

```python
train = df.iloc[:210,].reset_index(drop = True)
test = df.iloc[210:,].reset_index(drop = True)
```

```python
display(train,test)
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(train.drop(columns = ['target']))
X
```

```python
y = train['target']
```

```python
s = sm.Logit(y,X)
ss= s.fit()
ss.summary()
```

```python
ss.params['age']
```

```python
import numpy as np
```

```python
np.exp(ss.params['age'])
```

## train으로 로지스틱 회귀 진행했을 경우 잔차 이탈도 (residual deviance)를 계산하라

```python
# https://blog.naver.com/parksehoon1971/222343640308
# sm.Logit 에선 잔파 이탈도를 구하기 어려워서 sm.GLM을 사용한다
# family=sm.families.Binomial() 옵션을 줍니다. 이 옵션을 주지 않으면 로지스틱 회귀분석이 실행되지 않습니다.
# 로지스틱 회귀모형 적합 (GLM 사용) -> 이항분포시 로지스틱회귀
```

```python
m = sm.GLM(y, X, family=sm.families.Binomial())
```

```python
mm = m.fit()
```

```python
mm.summary()
```

```python
mm.deviance
```

```python
display(ss.summary(),mm.summary()) # 두 결과 비교해보기
```

## train으로 로지스틱 회귀 진행했을 경우 로짓 우도값을 도출하라

```python
##summary를 통해 Log-Likelihood:-67.250
```

## test 데이터의 독립변수로 target 예측 후 오류율을 구하여라

```python
ss.predict(sm.add_constant(test.drop(columns = ['target'])))
```

```python
ss.predict(sm.add_constant(test.drop(columns = ['target']))) > 0.5
```

```python
pred = (ss.predict(sm.add_constant(test.drop(columns = ['target']))) > 0.5)*1
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test['target'],pred)
```

```python
# 오류율 : 1 - accuracy_score
1-accuracy_score(test['target'],pred)
```
